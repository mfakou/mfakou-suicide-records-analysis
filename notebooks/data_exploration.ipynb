```python
def min_max(x):
   min_index = suicides[x].idxmin()
   max_index = suicides[x].idxmax()

   min = pd.DataFrame(suicides.loc[min_index,:])
   max = pd.DataFrame(suicides.loc[max_index,:])

   return pd.concat([min, max], axis=1)
```

```python
suicides['num_of_suicides'].replace(0, np.NAN, inplace=True)
mean_num = suicides['num_of_suicides'].mean()
suicides['num_of_suicides'] = suicides['num_of_suicides'].fillna(mean_num)
min_max('num_of_suicides')
```

```python
suicides['suicides/100k'].replace(0, np.NAN, inplace=True)
mean_num = suicides['suicides/100k'].mean()
suicides['suicides/100k'] = suicides['suicides/100k'].fillna(mean_num)
min_max('suicides/100k')
```
