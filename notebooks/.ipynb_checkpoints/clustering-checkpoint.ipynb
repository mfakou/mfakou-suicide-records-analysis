```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import SilhouetteVisualizer
import scipy.cluster.hierarchy as shc
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('../data/suicides.csv')
# grouped_by_country = data.groupby(['country', 'year', 'sex'], as_index=False)[['suicides/100k', 'population', 'num_of_suicides']].sum()
# data = pd.merge(data, grouped_by_country, how='inner', on=['country', 'year', 'sex'])
# data.drop(['num_of_suicides_x', 'suicides/100k_x', 'population_x', 'generation', 'age'],
#           axis=1, inplace=True)
# data = data.drop_duplicates(subset=['country', 'year', 'sex'])
data
```

# Data Preprocessing

```python
to_convert = ['country', 'year', 'sex', 'age', 'continent', 'generation']
data[to_convert] = data[to_convert].astype('category')
data[to_convert] = pd.DataFrame((pd.factorize(data[to_convert].values.ravel())[0] + 1).reshape(data[to_convert].shape),
                 columns=to_convert)

to_scale = ['num_of_suicides', 'population', 'suicides/100k', 'gdp_for_year', 'gdp_per_capita']
scaler = StandardScaler()
data[to_scale]= scaler.fit_transform(data[to_scale])
```

Principal Component Analysis

```python
pca = PCA()
sc_data = pca.fit_transform(data)
plt.plot(range(11), np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
pca = PCA(n_components=2)
sc_data = pca.fit_transform(data)
```

# Silhouette analysis and K-means application

```python
fig, ax = plt.subplots(3, 2, figsize=(15, 8))
for i in [2, 3, 4, 5, 6, 7]:
  model = KMeans(n_clusters=i, init='k-means++', random_state=42)
  q, mod = divmod(i, 2)
  visualizer = SilhouetteVisualizer(model, colors='yellowbrick', ax=ax[q-1][mod])
  visualizer.fit(sc_data)
```

```python
col = ['g', 'r', 'c', 'm', 'b', 'y']
for i in range(2, 7):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42) 
  kmeans.fit_predict(sc_data)
  df_pca = pd.concat([data, pd.DataFrame(sc_data)], axis=1)
  df_pca.columns.values[-2: ] = ['Component 1', 'Component 2']
  df_pca['labels'] = kmeans.labels_
  df_pca['labels'].unique()
  plt.figure(figsize = (10, 8))
  sns.scatterplot(x=df_pca['Component 1'], y=df_pca['Component 2'], hue=df_pca['labels'], palette=col[:i])
  plt.title('number of clusters = ' + str(i))
  plt.show()
  plt.clf()
```

# Dendrograms' construction and Agglomerative clustering application 

```python
for linkage_method in ['ward', 'complete', 'average']:
  plt.figure(figsize=(20, 10))
  plt.title('linkage = ' + linkage_method)
  # ward, complete, average, single
  shc.dendrogram(shc.linkage(sc_data, method=linkage_method))
  plt.show()
```

```python
col = ['g', 'r', 'c', 'm', 'b', 'y']
for i in range(2, 7):
  model = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
  model.fit_predict(sc_data)
  df_pca_agg = pd.concat([data, pd.DataFrame(sc_data)], axis=1)
  df_pca_agg.columns.values[-2: ] = ['Component 1', 'Component 2']
  df_pca_agg['labels'] = model.labels_
  plt.figure(figsize = (10, 8))
  sns.scatterplot(x=df_pca_agg['Component 1'], y=df_pca_agg['Component 2'], hue=df_pca_agg['labels'], palette=col[:i])
  plt.title('number of clusters = ' + str(i))
  plt.show()
  plt.clf()
```
